```
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv'

apple = pd.read_csv(url)
print(apple.head(5))
apple.info()
apple.Date = pd.to_datetime(apple.Date)
apple.dtypes
apple = apple.set_index('Date')
apple.head(2)
apple.groupby('Date').size().sort_values(ascending = False)
apple.index.is_unique
apple = apple.sort_index(ascending = True)
apple.head(5)
apple = apple.resample('BM')
print(apple.head(5))
#What is the difference in days between the first day and the oldest
result = (apple.index.max() - apple.index.min()).days
print(result)

apple_months = apple.resample('M').mean()
len(apple_months)

#Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches
check = apple['Adj Close'].plot(title = "Apple Stock")
```
